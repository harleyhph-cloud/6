## Ejercicio

Escribir un programa que pida al usuario una palabra y la muestre 10 veces por pantalla.

## Solución

```python
word = input("Introduce una palabra: ")
for i in range(10):
    print(word)
```

### Solución con bucle `while`

```python
word_while = input("Introduce una palabra para el bucle while: ")
count = 0
while count < 10:
    print(word_while)
    count += 1
```
